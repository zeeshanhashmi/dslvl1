```python
def fetchDataFromURL(url,dataIndex):
    import pandas as pd
    data = {'header':[],'data':[]}
    scrappedData = pd.read_html(url)
    if dataIndex >= len(scrappedData):
        print(f'Can\'t find index no. {dataIndex} loaded dataset.\nPlease use data index number between 0 to {dataIndex-1}')
    else:
        dataFrame = scrappedData[dataIndex]
        data["header"] = list(dataFrame.head().columns)
        data["data"] = dataFrame.to_numpy().tolist()
    return data
```

```python
url = "https://stats.espncricinfo.com/ci/content/records/284262.html"
data = fetchDataFromURL(url,0)
```

```python
def freq_table_st (data, index, st_index):

    unique_st = []
    freq_tst = {}
    
    for row in data:
        chkData = row[st_index]
        if chkData not in unique_st:
            unique_st.append(chkData)

    for st in unique_st:
        statsDict = {}

        for row in data:
            key = row[index]
            chkData = row[st_index]
            if chkData == st:
                if key in statsDict.keys():
                    statsDict[key] = statsDict[key]+1 
                else:
                    statsDict[key] = 1

                freq_tst[chkData] = statsDict
    return freq_tst
```

```python
ft = freq_table_st(data["data"], 1, 0)
```
